```python
# Graph Representation
graph = {0:[1,2],1:[2],2:[0,3],3:[3]}
```

```python
# BFS
from collections import deque
def bfs(graph,start):
    visited=set()
    q=deque([start])
    while q:
        n=q.popleft()
        if n not in visited:
            print(n,end=" ")
            visited.add(n)
            q.extend(graph[n])
bfs(graph,0)
```

```python
# DFS
def dfs(graph,node,visited=None):
    if visited is None: visited=set()
    if node not in visited:
        print(node,end=" ")
        visited.add(node)
        for n in graph[node]:
            dfs(graph,n,visited)
dfs(graph,0)
```

```python
# Topological Sort
from collections import defaultdict,deque
def topo(V,edges):
    g=defaultdict(list)
    indeg=[0]*V
    for u,v in edges:
        g[u].append(v); indeg[v]+=1
    q=deque([i for i in range(V) if indeg[i]==0])
    res=[]
    while q:
        n=q.popleft()
        res.append(n)
        for x in g[n]:
            indeg[x]-=1
            if indeg[x]==0: q.append(x)
    return res
print(topo(6,[(5,2),(5,0),(4,0),(4,1),(2,3),(3,1)]))
```

```python
# Dijkstra
import heapq
graph_w={0:[(1,4),(2,1)],1:[(3,1)],2:[(1,2),(3,5)],3:[]}
def dijkstra(g,s):
    dist={i:float('inf') for i in g}
    dist[s]=0
    pq=[(0,s)]
    while pq:
        d,n=heapq.heappop(pq)
        for v,w in g[n]:
            if d+w<dist[v]:
                dist[v]=d+w
                heapq.heappush(pq,(d+w,v))
    return dist
print(dijkstra(graph_w,0))
```

```python
# Bellman-Ford
def bellman(V,edges,src):
    dist=[float('inf')]*V
    dist[src]=0
    for _ in range(V-1):
        for u,v,w in edges:
            if dist[u]+w<dist[v]:
                dist[v]=dist[u]+w
    return dist
print(bellman(4,[(0,1,4),(0,2,1),(2,1,2),(1,3,1),(2,3,5)],0))
```

```python
# Prim MST
import heapq
def prim(g,start):
    vis=set()
    pq=[(0,start)]
    cost=0
    while pq:
        w,n=heapq.heappop(pq)
        if n not in vis:
            vis.add(n); cost+=w
            for v,w2 in g[n]:
                if v not in vis: heapq.heappush(pq,(w2,v))
    return cost
print(prim(graph_w,0))
```

```python
# Kruskal MST
def find(p,i):
    if p[i]==i: return i
    return find(p,p[i])

def kruskal(V,edges):
    p=list(range(V))
    edges=sorted(edges,key=lambda x:x[2])
    res=[]
    for u,v,w in edges:
        x=find(p,u); y=find(p,v)
        if x!=y:
            res.append((u,v,w))
            p[x]=y
    return res
print(kruskal(4,[(0,1,4),(0,2,1),(1,2,2),(1,3,1),(2,3,5)]))
```
